## 1. concat을 이용한 결합

### 1.1 내장문서

```python
?pd.concat
```

### 1.2 Series

```python
#Series
x = pd.Series(['A','B','C'], index=[1,2,3])
y = pd.Series(['D','E','F'], index=[4,5,6])
print(x); print(y)
```

```python
pd.concat([x,y])
```

### 1.3 DataFrame

```python
#Dataframe
import numpy as np
x = np.random.randint(10, size=(2,2))
y = np.random.randint(10, size=(2,2))
df1 = pd.DataFrame(x, columns=['A','B'], index=[1,2])
df2 = pd.DataFrame(y, columns=['A','B'], index=[1,2])
print(df1)
print(df2)
```

#### 결합과 방향 설정
* axis 설정에 따라 행 또는 열 방향으로 결합 가능

```python
# 행방향 결합
pd.concat([df1,df2])
# pd.concat([df1,df2], axis=0)
```

```python
# 열방향 결합
pd.concat([df1,df2], axis=1)
```

* 위 결과를 보면 기존 인덱스가 유지되는 것을 볼 수 있음
* 기존 인덱스를 무시하고 새로운 정수 인덱스 생성하고 싶을 경우 ignore_index 사용

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
pd.concat([df1,df2], axis=1, ignore_index=True)
```

* 다중 인덱스 키를 추가하고 싶은 경우, keys 옵션 사용

```python
pd.concat([df1,df2], keys=['2020','2021'])
```

* 열을 병합할 때 사용 가능한 join 옵션

```python
x = np.random.randint(10,size=(2,3))
y = np.random.randint(10,size=(2,2))
df1 = pd.DataFrame(x, columns=['A','B','C'], index=[1,2])
df2 = pd.DataFrame(y, columns=['C','D'], index=[1,2])
print(df1); print(df2)
```

```python
pd.concat([df1,df2])
```

```python
# 열의 합집합
pd.concat([df1,df2], join='outer')
```

```python
# 열의 교집합
pd.concat([df1,df2], join='inner')
```

```python
# 반환할 열 지정
pd.concat([df1,df2], join='outer')[df1.columns]
```

```python
# 3개 이상의 데이터프레임 연결
z = np.random.randint(10, size=(3,2))
df3 = pd.DataFrame(z, columns=['A','D'], index=[1,2,3])
pd.concat([df1,df2,df3], join='outer')
```

* concat() 대신 append() 메서드 사용도 가능, 이때는 결합된 데이터를 새로운 객체로 생성

```python
tmp = df1.append(df2)
tmp.append(df3)
```

## 2. merge를 이용한 데이터베이스 스타일 결합

### 2.1 유형 
#### 1) 일대일(one-to-one) 조인
* 자동으로 공통의 열을 찾아 조인

```python
df1 = pd.DataFrame({'employee':['Bob','Jake','Lisa','Sue'],
                    'group':['Accounting','Engineering','Engineering','HR']})
df2 = pd.DataFrame({'employee':['Lisa', 'Bob','Jake','Sue'],
                    'hire_date':[2004, 2008, 2012, 2014]})
print(df1); print(df2)
```

```python
df3 = pd.merge(df1,df2)
df3
```

#### 2) 다대일(many-to-one) 조인
* 키가 되는 열에 중복된 항목이 존재하는 경우의 조인
* 중복된 항목에 조인하는 값이 여러번 반복해서 등장

```python
df4 = pd.DataFrame({'group':['Accounting','Engineering','HR'],
                    'supervisor':['Carly','Guido','Steve']})
print(df3); print(df4)
```

```python
pd.merge(df3,df4)
```

#### 3) 다대다(many-to-many) 조인 
* 키가 되는 열 모두 중복 항목이 존재할 경우

```python
df5 = pd.DataFrame({'group':['Accounting','Accounting','Engineering','Engineering','HR','HR'],
                    'skills':['math','spreadsheets','coding','linux','spreadsheets','organization']})
print(df1); print(df5)
```

```python
pd.merge(df1,df5)
```

### 2.2 방향
* how 옵셥 사용
    * inner : 내부조인(교집합)
    * outer : 외부조인(합집합), 누락된 값은 NA로 채움
    * left / right : 특정 방향을 기준으로 조인



```python
df7 = pd.DataFrame({'name':['Peter','Paul','Mary'],
                    'food':['fish','beans','bread']})
df8 = pd.DataFrame({'name':['Mary','Joseph'],
                    'drink':['wine','drink']})
print(df7); print(df8)
```

```python
pd.merge(df7,df8)
```

```python
pd.merge(df7,df8, how='inner')
```

```python
pd.merge(df7,df8, how='outer')
```

```python
pd.merge(df7,df8, how='left')
```

```python
pd.merge(df7,df8, how='right', on='name')
```

### 2.3 병합할 컬럼 지정 방법
* 병합할 두 데이터셋의 열 이름을 on을 사용해 명시적으로 지정

```python
print(df1); print(df2); 
```

```python
pd.merge(df1,df2, on='employee')
```

* 두 데이터셋의 열 이름이 일치하지 않는 경우는 left_on과 right_on 사용

```python
df6 = pd.DataFrame({'name':['Bob','Jake','Lisa','Sue'],
                    'salary':[70000,80000,120000,90000]})
print(df1); print(df6)
```

```python
pd.merge(df1,df6, left_on='employee',right_on='name')
```

* 중복된 열을 dorp() 메서드로 삭제

```python
pd.merge(df1,df6, left_on='employee',right_on='name').drop('name',axis=1)
```

### 2.4 병합 시 열 이름이 겹치는 경우 처리
- on 파라미터에 키 값으로 사용할 변수(컬럼) 지정

```python
df9 = pd.DataFrame({'name':['Bob','Jake','Lisa','Sue'],
                    'rank':[1,2,3,4]})
df10 = pd.DataFrame({'name':['Bob','Jake','Lisa','Sue'],
                    'rank':[3,1,4,2]})
print(df9); print(df10)
```

```python
pd.merge(df9,df10,on='name')
```

```python
pd.merge(df9,df10,on='name', suffixes=['_L','_R'])
```

### 2.5 인덱스를 활용한 병합 방법

## 4. 피벗 테이블
GroupBy의 다차원 버전

```python
# seaborn library : titanic data
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#성별에 따른 생존률 GroupBy
titanic.groupby('sex')[['survived']].mean()
```

```python
# Drill Down 1, 성별과 좌석 등급별 생존률
titanic.groupby(['sex','class'])['survived'].mean().unstack()
```

- 위와같은 다차원 집계를 간결하게 처리할 수 있는 pivot_table

```python
titanic.pivot_table('survived',index='sex',columns='class')
```

```python
# Drill Down 2, 연령을 추가하여 살펴보기
age = pd.cut(titanic['age'],[0,18,80])
titanic.pivot_table('survived',index=['sex',age],columns='class')
```

```python
# Drill Down 3, 지불 비용을 추가하여 살펴보기
fare =  pd.qcut(titanic['fare'],2)
titanic.pivot_table('survived',['sex',age],[fare,'class'])
```

- aggfunc 옵션은 어떤 유형의 집계를 적용할지 선택, 기본은 평균
  - 'sum','count','min','max' 등의 문자열
  - 또는 집계를 구현하는 함수로 표현, np.sum(), min(), sum() 등
  - 딕셔너리를 이용해 열을 원하는 집계 방식으로 설정 가능

```python
titanic.pivot_table(index='sex',
                    columns='class',
                    aggfunc={'survived':'sum', 'fare':'mean'})
```

- margin 옵션 : 컬럼에 상관없이 해당 그룹의 집계

```python
titanic.pivot_table('survived', index='sex', columns='class', margins=True)
```
