### [확인학습 2] 다음 두 데이터프레임을 이용해 문제를 해결하시오.

```python
data1_dic = {
    'id':['0','1','2','3','4','6','8','11','12','13'],
    'city':['Seoul','Pusan','Daegu','Gangneung','Seoul','Seoul','Pusan','Daegu','Gangneung','Seoul'],
    'birth_year':[1990,1989,1992,1997,1982,1991,1988,1990,1995,1981],
    'name':['Junho','Heejin','Mijung','Minho','Steeve','Mina','Sumi','Minsu','Jinhee','Daeho']
}
data2_dic = {
    'id':['70','80','90','120','150'],
    'city':['Ilsan','Gunpo','Seoul','Changwon','Jeju'],
    'birth_year':[1980,1999,1995,1994,1994],
    'name':['Jinhee','Yeongho','Jongho','Yeonghee','Hyejin']
}
data1=pd.DataFrame(data1_dic)
data2=pd.DataFrame(data2_dic)
```

1. 두 데이터를 INNER JOIN 하시오.

```python
pd.merge(data1,data2, how = 'inner')
```

2. 두 데이터를 FULL JOIN 하시오.

```python
pd.merge(data1,data2, how = 'outer')
```

3. 두 데이터를 수직방향으로 결합하시오.

```python
pd.concat([data1,data2], axis = 0)
```

### [확인학습 3]

1. 수학성적 데이터 student-mat.csv를 읽어 들여, 연령(age)에 2를 곱한 새로운 컬럼을 마지막 열에 추가하시오.

```python
df = pd.read_csv('student-mat.csv', ';')
age_mul_2 = pd.Series(df['age']*2,name = 'age_mul_2')
df = pd.concat([df,age_mul_2],axis = 1)
df
```

2. absences 컬럼을 세 개의 구간으로 나누고 각 구간별 학생수를 계산하시오.(구간 분할 간격 absences_bin = [0,1,5,100] )

```python
absences_bin = [0,1,5,100]
val = pd.cut(df['absences'],absences_bin)
df.pivot_table('absences',index=val,aggfunc={'absences':'sum'})
```

3. absences 컬럼을 qcut 함수로 세 개의 구간으로 분할하시오.

```python
q_cut = pd.qcut(df['absences'],2)
q_cut
```

4. 학교(school) 변수를 기준으로 각 학교의 G1 평균 점수를 구하시오.

```python
df.pivot_table(index = 'school', aggfunc = {'G1':'mean'})
```

5. 학교(school)와 성별(sex)를 기준으로 G1,G2,G3 평균 점수를 구하시오.

```python
df.pivot_table(index =['school','sex'],aggfunc = {'G1':'mean','G2':'mean','G3':'mean'})
```

6. 학교(school)와 성별(sex)를 기준으로 G1,G2,G3 최댓값을 구하시오.

```python
df.pivot_table(index =['school','sex'],aggfunc = {'G1':'max','G2':'max','G3':'max'})
```
